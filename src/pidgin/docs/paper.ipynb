```python
    def show(object):
        path = pathlib.Path(object.__file__)
        return nbconvert.get_exporter('markdown')(exclude_input='.md' in path.suffixes).from_filename(path)[0]
    
    IPython.get_ipython().display_formatter.mimebundle_formatter.type_printers.pop(types.ModuleType, None)
    IPython.get_ipython().display_formatter.formatters['text/markdown'].for_type(types.ModuleType, show)
```
